## 1. Introduction to Stochastic Calculus

Classical calculus (based on Riemann integration) is not suitable for integrating with respect to stochastic processes like Brownian motion. This is because the paths of Brownian motion are of unbounded variation, meaning they are too "wiggly" for the Riemann-Stieltjes integral to be well-defined.

Stochastic calculus, particularly ItÃ´ calculus, provides a framework for integrating with respect to such processes.

## 2. The ItÃ´ Integral

The ItÃ´ integral of a process H_t with respect to Brownian motion W_t is denoted as:
$I(t) = \int_0^t H_s \, dW_s$

It is defined as the limit in probability of a sum:
$\sum H_{t_{i-1}} (W_{t_i} - W_{t_{i-1}})$
where the partition of the interval [0, t] becomes finer.

A key feature of the ItÃ´ integral is that the integrand H_t is evaluated at the left endpoint of the subintervals. This leads to some non-intuitive results, such as the ItÃ´ isometry.

### Properties of the ItÃ´ Integral
1. **Linearity:** $\int (aH_1 + bH_2) dW = a\int H_1 dW + b\int H_2 dW$
2. **Martingale property:** The ItÃ´ integral is a martingale.
3. **ItÃ´ isometry:** $E\left[\left(\int H dW\right)^2\right] = E\left[\int H^2 dt\right]$

## 3. ItÃ´ Processes

An **ItÃ´ process** is a stochastic process X_t that can be written in the form:
$dX_t = a(t, X_t) dt + b(t, X_t) dW_t$
or in integral form:
$X_t = X_0 + \int_0^t a(s, X_s) ds + \int_0^t b(s, X_s) dW_s$

Here, $a(t, X_t)$ is the drift term and $b(t, X_t)$ is the diffusion term.
