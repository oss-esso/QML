# Module 3: Stochastic Calculus - The Itô Integral

This notebook introduces the Itô integral, a fundamental concept in stochastic calculus.

## 1. Introduction to Stochastic Calculus

Classical calculus (based on Riemann integration) is not suitable for integrating with respect to stochastic processes like Brownian motion. This is because the paths of Brownian motion are of unbounded variation, meaning they are too "wiggly" for the Riemann-Stieltjes integral to be well-defined.

Stochastic calculus, particularly Itô calculus, provides a framework for integrating with respect to such processes.

## 2. The Itô Integral

The Itô integral of a process H_t with respect to Brownian motion W_t is denoted as:
I(t) = ∫_0^t H_s dW_s

It is defined as the limit in probability of a sum:
Σ H_{t_{i-1}} (W_{t_i} - W_{t_{i-1}})where the partition of the interval [0, t] becomes finer.

A key feature of the Itô integral is that the integrand H_t is evaluated at the left endpoint of the subintervals. This leads to some non-intuitive results, such as the Itô isometry.

### Properties of the Itô Integral
1. **Linearity:** ∫ (aH_1 + bH_2) dW = a∫ H_1 dW + b∫ H_2 dW
2. **Martingale property:** The Itô integral is a martingale.
3. **Itô isometry:** E[(∫ H dW)^2] = E[∫ H^2 dt]

## 3. Itô Processes

An **Itô process** is a stochastic process X_t that can be written in the form:
dX_t = a(t, X_t) dt + b(t, X_t) dW_t
or in integral form:
X_t = X_0 + ∫_0^t a(s, X_s) ds + ∫_0^t b(s, X_s) dW_s

Here, a(t, X_t) is the drift term and b(t, X_t) is the diffusion term.
